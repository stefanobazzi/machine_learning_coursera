### Online Learning

The online learning setting allows us to model problems where we have a continuous flood or a continuous stream of data coming in and we would like an algorithm to learn from that. Specifically, if you have a continuous stream of data generated by a continuous stream of users coming to your website, what you can do is sometimes use an online learning algorithm to learn user preferences from the stream of data and use that to optimize some of the decisions on your website.

Repeat forever {      
&nbsp;&nbsp;&nbsp;&nbsp; Get (x, y) corresponding to user    
&nbsp;&nbsp;&nbsp;&nbsp; Update $\theta$ using (x, y):      
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ \theta_j := \theta_j - \alpha(h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}$       ($ j = 0, \dots, n$)    
}     

#### Online learning example

Product search (learning to search)
User searches fir "Android phone 1080p camera"    
Have 100 phones in store. Will return 10 results.    
$x$ = features of phone, how many words in user query match name of phone, how many words in query match description of phone, etc.   
$y = 1$ if user clicks on link $y=0$ otherwise    
Learn $p(y=1 | x;0)$     
Use to show user the 10 phones they're most likey to click on.     
Other examples: Chooising special offers to show user; customized selection of new articles; product recommendation;...



### Map Reduce and Data Parallelism

#### Mapreduce

m = 400, Batch gradient descent $ \theta_j := \theta_j - \alpha\frac{1}{400}\displaystyle\sum_{i=1}^{400}(h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}$      

Split dataset into four datasets

Machine 1: Use $ ((x^{(1)}, y^{(1)}), \dots, (x^{(100)}, y^{(100)})) $     
&nbsp;&nbsp;&nbsp;&nbsp; $ temp_j^{(1)} = \displaystyle\sum_{i=1}^{100}(h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}$ 

Machine 2: Use $ ((x^{(101)}, y^{(101)}), \dots, (x^{(200)}, y^{(200)})) $     
&nbsp;&nbsp;&nbsp;&nbsp; $ temp_j^{(2)} = \displaystyle\sum_{i=101}^{200}(h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}$ 

Machine 3: Use $ ((x^{(201)}, y^{(201)}),  \dots, (x^{(300)}, y^{(300)})) $     
&nbsp;&nbsp;&nbsp;&nbsp; $ temp_j^{(3)} = \displaystyle\sum_{i=201}^{300}(h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}$ 

Machine 4: Use $ ((x^{(301)}, y^{(301)}), \dots, (x^{(400)}, y^{(400)})) $     
&nbsp;&nbsp;&nbsp;&nbsp; $ temp_j^{(4)} = \displaystyle\sum_{i=301}^{400}(h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}$ 

Combine:  

$ \theta_j := \theta_j - \alpha\frac{1}{400}(temp_j^{(1)} + temp_j^{(2)} + temp_j^{(3)} + temp_j^{(4)})$

#### Map-reduce and summation over the training set

Many learning algorithms can be expressed as computing sums of functions over the training set.

E.g for advancede optimization, with logistic regression, need:

Split both sums over $i$:     
$ J(\theta) =  -\frac{1}{m}\sum_{i=0}^{m}[y^{(i)}\log(h_\theta(x^{(i)})) + (1 - y^{(i)})\log(1 - h_\theta(x^{(i)}))] $     
$ \theta_j := \theta_j - \alpha\sum_{i=0}^{m}(h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}$
