### Optimization Objective

Alternative view of logistic regression   

$h_{\theta}(x) = \frac{1}{1 + e^{-\theta^Tx}}$

if $ y = 1 $, we want $h_{\theta}(x) \approx 1$ , $\theta^tx \gg 0 $     
if $ y = 0 $, we want $h_{\theta}(x) \approx 0$ , $\theta^tx \ll 0 $


Cost example $ - y\log(h_\theta(x)) + (1 - y)\log(1 - h_\theta(x)) =
- y\log(\frac{1}{1 + e^{-\theta^Tx}}) + (1 - y)\log(1 - \frac{1}{1 + e^{-\theta^Tx}})$

if  $ y = 1 $ (want $\theta^Tx \gg 0 $): - $\log(\frac{1}{1 + e^{-z}}) $  
if  $ y = 0 $ (want $\theta^Tx \ll 0 $): - $\log(1 - \frac{1}{1 + e^{-z}}) $



##### Logistic Regression:        
$ \min_\limits{\theta}  = \frac{1}{m}\sum_{i=1}^{m}[y^{(i)}(-\log(h_\theta(x^{(i)}))) + (1 - y^{(i)})(-\log(1 - h_\theta(x^{(i)})))] + \frac{\lambda}{2m}\sum_{i=1}^{n}\theta_j^2 $

put:

$ - \log(h_\theta(x^{(i)})) = cost_1(\theta^Tx^{(i)}) $   
$ - \log(1 - h_\theta(x^{(i)})) = cost_0(\theta^Tx^{(i)}) $   


$ \min_\limits{\theta}  = \frac{1}{m}\sum_{i=1}^{m}y^{(i)}cost_1(\theta^Tx^{(i)}) + (1 - y^{(i)})cost_0(\theta^Tx^{(i)}) + \frac{\lambda}{2m}\sum_{i=1}^{n}\theta_j^2 $

- get rid of $m$  
- instead of $\lambda$ will be a parameter $C$ to multiply the first term

##### Suppert  vectorm machine

Cost function:   
$ \min_\limits{\theta}  = C\sum_{i=1}^{m}[y^{(i)}cost_1(\theta^Tx^{(i)}) + (1 - y^{(i)})cost_0(\theta^Tx^{(i)})] + \frac{1}{2}\sum_{i=1}^{n}\theta_j^2 $

Hypothesis:    

$ h_\theta(x) = 
\begin{cases}
1 & \text{if } \theta^Tx \geq 0 \\
0 & \text{otherwise} \\
\end{cases}$

### Large margin Intuition

Cost function:   
$ \min_\limits{\theta}  = C\sum_{i=1}^{m}[y^{(i)}cost_1(\theta^Tx^{(i)}) + (1 - y^{(i)})cost_0(\theta^Tx^{(i)})] + \frac{1}{2}\sum_{i=1}^{n}\theta_j^2 $

if  $ y = 1 $ want $\theta^Tx \geq 1 $  (not just $\geq 0$)    
if  $ y = 0 $ want $\theta^Tx \leq -1 $  (not just $< 0$)

Whenever $ y^{(i)} = 1 $ :   
$\theta^Tx \geq 1 $  ->   $ \min\frac{1}{2}\sum_{i=1}^{n}\theta_j^2 $

### Mathematics Behind Large Margin Classification

##### Vector inner product 

$ u = 
\begin{bmatrix}
u_1 \\
u_2
\end{bmatrix}$
$ v = 
\begin{bmatrix}
v_1 \\
v_2
\end{bmatrix}$

$\|u\| = $ length of vector $u$ = $ \sqrt{u^2 + v^2}$

$u^Tv = u_1v_1 + u_2v_2$

$p$ = length of projection of $v$ into $u$   
$u^Tv = p\|u\|$


##### SVM Decision Boundary

Semplification $\theta_0 = 0$, $n = 2$ 

$ \min\frac{1}{2}\sum_{i=1}^{n}\theta_j^2 = \frac{1}{2}(\theta_1^{2} + \theta_2^{2})  = \frac{1}{2}\|\theta\|^2$

$\theta^Tx^{(i)} = p^{(i)}\|\theta\|$

$\theta^Tx^{(i)} \geq  1 $ if  $ y^{(i)} = 1 $   -> $p^{(i)}\|\theta\| \geq 1 $ if  $ y^{(i)} = 1 $     
$\theta^Tx^{(i)} \leq -1 $ if  $ y^{(i)} = 0 $   -> $p^{(i)}\|\theta\| \leq -1 $ if  $ y^{(i)} = 0 $
